```python
# Converting Images to array 

def convert_image_to_array(image_dir):
    try:
        image = cv2.imread(image_dir)
        if image is not None :
            image = cv2.resize(image, (256, 256))  
            return img_to_array(image)
        else :
            return np.array([])
    except Exception as e:
        print(f"Error : {e}")
        return None
```

```python
dataset_path = '/Users/surajsamal/Desktop/untitled folder/Data/Plant_images_pianalytix'

# Classes
all_labels = ['Corn-Common_rust', 'Potato-Early_blight', 'Tomato-Bacterial_spot']
binary_labels = [0, 1, 2]

# Lists for image arrays and labels
image_list, label_list = [], []

# Function to convert image to array
def convert_image_to_array(image_path, size=(128, 128)):
    try:
        image = Image.open(image_path).convert('RGB')
        image = image.resize(size)
        return np.array(image)
    except Exception as e:
        print(f"❌ Failed to convert {image_path}: {e}")
        return None

# Read images and labels
for idx, directory in enumerate(all_labels):
    full_dir = os.path.join(dataset_path, directory)
    if not os.path.isdir(full_dir):
        continue

    plant_image_list = [f for f in listdir(full_dir) if not f.startswith('.')]  # skip .DS_Store or hidden
    for file in plant_image_list:
        image_path = os.path.join(full_dir, file)
        image_array = convert_image_to_array(image_path)
        if image_array is not None:
            image_list.append(image_array)
            label_list.append(binary_labels[idx])
```

```python
# Visualize the number of classes count

label_counts = pd.DataFrame(label_list).value_counts()
label_counts.head()
```

```python
print("Number of images loaded:", len(image_list))
```

```python
label_list = np.array(label_list)
label_list.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(image_list, label_list, test_size=0.2, random_state = 10) 
```

```python
x_train = np.array(x_train, dtype=np.float16) / 225.0
x_test = np.array(x_test, dtype=np.float16) / 225.0
x_train = x_train.reshape(-1, 256, 256, 3)
x_test = x_test.reshape(-1, 256, 256, 3)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding = "same",input_shape = (256, 256, 3), activation = "relu"))
model.add(MaxPooling2D(pool_size = (3, 3)))
model.add(Conv2D(16, (3, 3), padding = "same", activation = "relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())
model.add(Dense(8, activation = "relu"))
model.add(Dense(3, activation = "softmax"))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = Adam(0.0001), metrics = ['accuracy'])
```

```python
import numpy as np

X = np.array(image_list)
y = np.array(label_list)
```

```python
x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Training the model

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(len(np.unique(y_train)), activation='softmax')  # auto handles 3 classes
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
x_train = x_train / 255.0
x_val = x_val / 255.0
```

```python
epochs = 50
batch_size = 128

history = model.fit(
    x_train, y_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(x_val, y_val)
)
```

```python
model.save("C:\Plant-Disease-Detection\Model\plant_disease_model.h5")
```

```python
plt.figure(figsize = (12, 5))
plt.plot(history.history['accuracy'], color = 'r')
plt.plot(history.history['val_accuracy'], color = 'b')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'val'])

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

y_pred = model.predict(x_val)
y_pred_labels = np.argmax(y_pred, axis=1)

cm = confusion_matrix(y_val, y_pred_labels)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=all_labels, yticklabels=all_labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print("Calculating model accuracy")
scores = model.evaluate(x_val, y_val)
print(f"Validation Accuracy: {scores[1] * 100:.2f}%")
```
